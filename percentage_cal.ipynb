```python
import numpy as np
import time
import os
import pandas as pd
import requests
import yfinance as yf
from datetime import datetime, date, timedelta
```

```python
def get_50nsedata():
    data = pd.read_csv('Option_stocks.txt')
    nse_50data = data['Symbol'].to_list()
    nse_50data = nse_50data[1:]
    return nse_50data
```

```python
'''
Download data from Yahoo finance based on stock name and date, interval
'''
def get_histdata(ticker, startdate, enddate, interval):
    out_dict = {'Symbol' : [], 'Percentage' : []};
    print('startdate :', startdate)
    print('enddate :', enddate)

    tickk = ["{}.ns".format(tick) for tick in ticker]
    for line in tickk:
        symbol = line.split(",")[0]
        data = yf.download(symbol, start=startdate, end=enddate, interval=interval)
        data = data[(data[['Volume']] != 0).all(axis=1)]
        data.to_csv('C:/Users/admin/Desktop/option chain analysis/datasets/options/{}.csv'.format(symbol))
        percent_cal = percentage_cal(data,symbol,out_dict)
    return percent_cal
```

```python
def percentage_cal(data,symbol,out_dict):
    try:
        close = round(float(data['Close'].iloc[-1]),2)
        adj_close = round(float(data['Close'].iloc[0]),2)
        diff = (close - adj_close)
        out_prcnt = ((diff/adj_close)*100)
        out_prcnt = round(out_prcnt,2)
        out_dict["Symbol"].append(symbol)
        out_dict["Percentage"].append(out_prcnt)
    except EOFError:
        print('End of file')
    return out_dict
```

```python
def Gainers(g1,Gain_stk):
    for checkk in g1:
        rate = pd.read_csv('C://Users//admin//Desktop//option chain analysis//datasets//options//{}.csv'.format(checkk))
        if (rate['Close'].iloc[-1]).item() > (rate['Open'].iloc[-1]).item() and (rate['Close'].iloc[-2]).item() > (rate['Open'].iloc[-2]).item() and (rate['Close'].iloc[-3]).item() > (rate['Open'].iloc[-3]).item():
            Gain_stk.append(checkk)
        else:
            pass
    return Gain_stk
```

```python
def Lossers(l1,Loss_stk):
    for checkk in l1:
        rate = pd.read_csv('C:/Users/admin/Desktop/option chain analysis/datasets/options/{}.csv'.format(checkk))
        if (rate['Close'].iloc[-1]).item() < (rate['Open'].iloc[-1]).item() and (rate['Close'].iloc[-2]).item() < (rate['Open'].iloc[-2]).item() and (rate['Close'].iloc[-3]).item() < (rate['Open'].iloc[-3]).item():
            Loss_stk.append(checkk)
        else:
            pass
    return Loss_stk
```

```python
def send_mail(percent_cal):
    prcnt_srt = pd.DataFrame(percent_cal)
    Gain_ten = prcnt_srt.sort_values(by=['Percentage'], ascending=False).head(10)
    Less_ten = prcnt_srt.sort_values(by=['Percentage']).head(10)
    return Gain_ten,Less_ten
```

```python
if __name__ == '__main__':
    
    Nifty_data = get_50nsedata()
    date = int(input('enter the date : '))
    month = int(input('enter the Month : '))
    startdate = datetime(2022,month,date,15,20,0)
    print(startdate)
    enddate = datetime.today() + timedelta(hours=11) + timedelta(minutes=30)
    print(enddate)
    interval = '5m'
    out = get_histdata(Nifty_data, startdate, enddate, interval)
    Gain_ten,Less_ten = send_mail(out)
    print('Top 10 gainers : \n',Gain_ten)
    print('Top 10 Lossers : \n',Less_ten)
    g1 = Gain_ten['Symbol'].to_list()
    l1 = Less_ten['Symbol'].to_list()
    Gain_stk = []
    Loss_stk = []
    gain_out = Gainers(g1,Gain_stk)
    loss_out = Lossers(l1,Loss_stk)
    print('Top 3 gainers stock : ',gain_out)
    print('Top 3 looser stock : ',loss_out)
```
